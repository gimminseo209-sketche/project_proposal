```python
import random
import argparse


def generate_lotto_numbers(n=6, min_num=1, max_num=45):
    if max_num - min_num + 1 < n:
        raise ValueError("범위가 너무 작습니다.")
    return sorted(random.sample(range(min_num, max_num + 1), n))


def main():
    parser = argparse.ArgumentParser(
        description="Lotto number generator (1~45, no duplicates)"
    )
    parser.add_argument("-n", "--count", type=int, default=6, help="뽑을 숫자 개수 (기본 6)")
    parser.add_argument("--min", type=int, default=1, help="최소값 (기본 1)")
    parser.add_argument("--max", type=int, default=45, help="최대값 (기본 45)")
    parser.add_argument("--seed", type=int, default=None, help="랜덤 시드")

    # ⭐ Jupyter/VS Code용 핵심 수정
    args, _ = parser.parse_known_args()

    if args.seed is not None:
        random.seed(args.seed)

    numbers = generate_lotto_numbers(args.count, args.min, args.max)
    print(f"이번 주 로또 번호: {numbers}")


if __name__ == "__main__":
    main()
```
